```python
import re

def getAccuracy(line):
  floats = re.findall("\d+\.\d+", line)
  try:
    return float(floats[2])
  except IndexError:
    pass 

def getAccuracies(output):
  accs = []
  for line in output.splitlines():
    if line != "Saving new best checkpoint...":
      accs.append(getAccuracy(line))

  return accs
```

```python
import re

def getLoss(line):
  floats = re.findall("\d+\.\d+", line)
  try:
    return float(floats[1])
  except IndexError:
    pass 

def getLosses(output):
  accs = []
  for line in output.splitlines():
    if line != "Saving new best checkpoint...":
      accs.append(getLoss(line))

  return accs
```
