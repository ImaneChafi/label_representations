# Helper Methods


```python
import re

def getAccuracy(line):
  floats = re.findall("\d+\.\d+", line)
  return float(floats[2])

def getAccuracies(output):
  accs = []
  for line in output.splitlines():
    if line != "Saving new best checkpoint...":
      accs.append(getAccuracy(line))

  return accs
```
